```python
import pandas as pd
```

```python
df =pd.read_csv("MELBOURNE_HOUSE_PRICES_LESS.csv")
```

```python
df.head(4)
```

```python
df.tail(4)
```

```python
#No of rows and columns
print(df.shape)
```

```python
#Drop problematic rows
df = df.dropna()
```

```python
print(df.shape)
```

```python
rooms =df["Rooms"]
```

```python
price =list(df["Price"])
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()


plt.hist(x = price,edgecolor = "black",bins=10)

plt.xlabel("price range")
plt.ylabel("Number of houses")
plt.title("Histogram of price")
plt.show
```

```python
rooms_count = df["Rooms"].value_counts()
```

```python
rooms_count = dict(rooms_count)
```

```python
a = rooms_count.keys()
b = rooms_count.values()
```

```python
#pie
fig = plt.figure()

plt.pie(x = b ,labels = a )
plt.legend()
plt.show
```

```python
regionname_count = df["Regionname"].value_counts()
```

```python
regionname_count = dict(regionname_count)
```

```python
a = list(regionname_count.keys())
b = list(regionname_count.values())
```

```python
#pie
fig = plt.figure(figsize=(10,10))

plt.pie(x = b ,labels = a )
plt.legend()
plt.show
```

```python
#barplot
#Type
type_count = df["Type"].value_counts()
type_count = dict(type_count)
print(type_count)
```

```python
a = list(type_count.keys())
b = list(type_count.values())
```

```python
fig = plt.figure(figsize=(5,5))
plt.bar(x = a, height = b)
plt.title("bargraph of type")
plt.show
```

```python
price = df["Price"]
price = list(price)
distance= df["Distance"]
distance = list(distance)
```

```python
fig = plt.figure(figsize=(10,10))
#LINEPLOT
plt.plot(distance,price,linestyle ="",marker = "*",
         markeredgecolor ="black",markerfacecolor ="purple")
plt.xlabel("distance--->")
plt.ylabel("price---->")
plt.title("Distance vs Price")
plt.show
```
